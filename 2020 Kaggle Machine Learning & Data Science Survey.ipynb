```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv(r"./Data/kaggle_survey_2020_responses.csv", low_memory=False)
raw.shape
```

```python
raw.head()
```

```python
question = raw.iloc[0] # 질문
question
```

```python
answer = raw.drop([0])
answer.head()
```

```python
answer.info()
```

```python
question["Q1"]
```

```python
Q1 = answer["Q1"].value_counts(normalize=True).sort_index() * 100
Q1.plot.bar().set_title(question["Q1"] + "  (%)")
```

```python
sns.countplot(data = answer, x="Q1", palette="Reds") # seaborn as sns
```

```python
Q1 = answer["Q1"].value_counts().sort_index()
sns.countplot(data=answer.sort_values("Q1"), x="Q1",
              palette="Blues_r").set_title(question["Q1"]) # Blues_r / r = color
```

```python
answer["Q2"].value_counts().plot.bar().set_title(question["Q2"][:20])
```

```python
answer["Q1"].value_counts().index
```

```python
def show_countplot_by_qno(qno, fsize=(10,6), order=None):
    if not order:
        order = answer[qno].value_counts().index # 기본값
    
    plt.figure(figsize=fsize)
    sns.countplot(data=answer, y=qno,
                 order=order).set_title(question[qno])
show_countplot_by_qno("Q3", fsize=(10,12))
```
